```python
import paddle
paddle.utils.run_check()
```

```python
from paddlespeech.cli.tts.infer import TTSExecutor
from IPython.display import Audio

tts_executor = TTSExecutor()

text = "今天天气十分不错，百度也能做语音合成。"
output_file = "./data/paddlespeech/fastspeech2_csmsc_pwgan_csmsc.wav"
tts_executor(text=text, output=output_file)
display(Audio(filename=output_file, autoplay=True))
```

```python
tts_executor = TTSExecutor()

text = "今天天气十分不错，Paddle Speech也能做语音合成。但是英文丢失了。"
output_file = "./data/paddlespeech/fastspeech2_csmsc_pwgan_csmsc_missing.wav"
tts_executor(text=text, output=output_file)
display(Audio(filename=output_file, autoplay=True))
```

```python
tts_executor = TTSExecutor()

text = "早上好, how are you? 百度Paddle Speech一样能做中英文混合的语音合成。"
output_file = "./data/paddlespeech/fastspeech2_male_pwgan_male.wav"
tts_executor(text=text, output=output_file,
             am="fastspeech2_male", voc="pwgan_male",
             lang="mix", spk_id=174)
display(Audio(filename=output_file, autoplay=True))
```

```python
tts_executor = TTSExecutor()

text = "早上好, how are you? 百度Paddle Speech一样能做中英文混合的语音合成。"
output_file = "./data/paddlespeech/fastspeech2_male_hifigan_male.wav"
tts_executor(text=text, output=output_file,
             am="fastspeech2_male", voc="hifigan_male",
             lang="mix", spk_id=174)
display(Audio(filename=output_file, autoplay=True))
```

```python
tts_executor = TTSExecutor()

text = "我们的声学模型使用了 Fast Speech Two, 声码器使用了 Parallel Wave GAN and Hifi GAN."
output_file = "./data/paddlespeech/fastspeech2_mix_pwgan_csmsc.wav"
tts_executor(text=text, output=output_file,
             am="fastspeech2_mix", voc="pwgan_csmsc",
             lang="mix", spk_id=175, use_onnx=True)
display(Audio(filename=output_file, autoplay=True))
```

```python
tts_executor = TTSExecutor()

text = "早上好, how are you? 百度Paddle Speech一样能做中英文混合的语音合成。"
output_file = "./data/paddlespeech/fastspeech2_mix_pwgan_aishell3.wav"
tts_executor(text=text, output=output_file,
             am="fastspeech2_mix", voc="pwgan_aishell3",
             lang="mix", spk_id=174, use_onnx=True)
display(Audio(filename=output_file, autoplay=True))
```
